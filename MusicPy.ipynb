```python
# valence/energy 
# Build a scatter plot for each data type
plt.scatter(top100_df["valence"], top100_df["tempo"], marker="*", s=95,facecolors="green", edgecolors="black",
             alpha=0.5, linewidth=0.5)

# Incorporate the other graph properties
plt.title("Tempo vs. Valence")
plt.ylabel("Tempo")
plt.xlabel("Valence")
plt.grid(True)

# Save the figure
plt.savefig("Resources/Tempo_vs_Valence.png")
```
